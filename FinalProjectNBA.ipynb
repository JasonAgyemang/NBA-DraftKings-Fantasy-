**Preliminary Model Setup (Only 8 Player Constraint)**

To initiate the process, I've created an empty model to grasp the mechanics of the problem, focusing solely on the 8-player constraint. Subsequently, I will integrate additional constraints and fully address the problem.

To ensure accessibility, I've stored the Draft Kings (DK) scoring rules from figure 1 in a dataframe since the excel file lacked this information. Following that, I established a dictionary for the scoring rules to facilitate easier reference during the analysis.






```python
# Scoring rules DataFrame
scoring_rules = pd.DataFrame({
    'Category': ['PTS', '3P', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'Double-Double', 'Triple-Double'],
    'Points': [1, 0.5, 1.25, 1.5, 2, 2, -0.5, 1.5, 3]
})
scoring_rules
```

```python
# Create a dictionary for easier lookup
scoring_dict = dict(zip(scoring_rules['Category'], scoring_rules['Points']))
```

**Model Setup and Decison Variables**

With all the requisite data in hand for solving problem 1, we can proceed to establish the model and define the decision variables. These variables, denoting the players selected, will be binary, with a value of 1 indicating selection and 0 indicating non-selection.






```python
# Define the Pyomo Concrete Model
model = pe.ConcreteModel()
```

```python
# Define decision variables
model.players = pe.Set(initialize=feb22_stats.index)
model.x = pe.Var(model.players, domain=pe.Binary)
```

**Define the Objective Function**

Our objective function aims to maximize the accumulation of DK points as outlined in the scoring rules. We aim to select players who can maximize these points based on their performance in the games they participated in on Feb 22.

```python
# Define the objective function using the scoring dictionary
model.obj = pe.Objective(
    expr=sum(
        scoring_dict[category] * feb22_stats[category][i] * model.x[i] if category not in ['Double-Double', 'Triple-Double']
        else scoring_dict[category] * (feb22_stats[category][i] == 'Yes') * model.x[i]
        for category in scoring_dict.keys()
        for i in model.players
    ),
    sense=pe.maximize
)
model.obj.pprint()
```

**Define Constraints**

At this stage, we'll begin by establishing the initial constraint of having precisely 8 players, as we're solely focused on setting up the model to grasp the mechanics of the problem.






```python
# Define the one constraint
model.total_players = pe.Constraint(expr=sum(model.x[i] for i in model.players) == 8)
```

**Solve the Model**

```python
# Solve the simple optimization model
opt = pe.SolverFactory('glpk')
result = opt.solve(model)
print(result.solver.status, result.solver.termination_condition)
```

**See Output**

Let's look at the output to see what the model looks like without any constraints except for picking the 8 players.

```python
# Display the selected lineup and total points
selected_lineup = [i for i in model.players if pe.value(model.x[i]) == 1]
total_points = sum(
    scoring_dict[category] * feb22_stats[category][i] if category not in ['Double-Double', 'Triple-Double']
    else scoring_dict[category] * (feb22_stats[category][i] == 'Yes')
    for category in scoring_dict.keys()
    for i in selected_lineup
)

print("Total DK Points: {:.2f}".format(total_points))
```

```python
# Extract selected lineup information
selected_lineup_data = feb22_stats.loc[selected_lineup]

# Display the selected lineup in a table
output_table = selected_lineup_data[['Player', 'PTS', '3P', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'Double-Double', 'Triple-Double', 'SalaryFeb']].copy()
output_table['DK_Points'] = sum(
    scoring_dict[category] * selected_lineup_data[category] if category not in ['Double-Double', 'Triple-Double']
    else scoring_dict[category] * (selected_lineup_data[category] == 'Yes')
    for category in scoring_dict.keys()
)

# Display the table
output_table
```

Although our lineup achieves a total DK points of 473.50 from the players selected for our roster, it does so without adhering to any constraints except for the limit of 8 players. Despite maximizing DK points from the Feb 22 games, this lineup blatantly disregards all other constraints.






## Problem 2

So now that we understand how the problem works and what we're solving for, let's fully solve problem 2.

**Model Setup and Decison Variables**

```python
# Define the Pyomo Concrete Model
model = pe.ConcreteModel()
```

```python
# Define decision variables
model.players = pe.Set(initialize=feb22_stats.index)
model.x = pe.Var(model.players, domain=pe.Binary)
for DV in model.component_objects(pe.Var):
    DV.pprint()
```

```python
# Define the objective function using the scoring dictionary
model.obj = pe.Objective(
    expr=sum(
        scoring_dict[category] * feb22_stats[category][i] * model.x[i] if category not in ['Double-Double', 'Triple-Double']
        else scoring_dict[category] * (feb22_stats[category][i] == 'Yes') * model.x[i]
        for category in scoring_dict.keys()
        for i in model.players
    ),
    sense=pe.maximize
)
model.obj.pprint()
```

**Define Constraints**


```python
# Define constraints
model.total_players = pe.Constraint(expr=sum(model.x[i] for i in model.players) == 8)
model.total_salary = pe.Constraint(expr=sum(feb22_stats['SalaryFeb'][i] * model.x[i] for i in model.players) <= 50000)
model.games_constraint = pe.Constraint(expr=sum(model.x[i] for i in model.players if feb22_stats['GameInfo'][i] != feb22_stats['GameInfo'][selected_lineup[0]]) >= 1)
# Constraint: At least one Point Guard (PG)
model.PG = pe.Constraint(expr=sum(model.x[i] for i in model.players if 'PG' in feb22_stats.loc[i, 'RosterPos'].split('/')) >= 2)
# Constraint: At least one Shooting Guard (SG)
model.SG = pe.Constraint(expr=sum(model.x[i] for i in model.players if 'SG' in feb22_stats.loc[i, 'RosterPos'].split('/')) >= 2)
# Constraint: At least one Small Forward (SF)
model.SF = pe.Constraint(expr=sum(model.x[i] for i in model.players if 'SF' in feb22_stats.loc[i, 'RosterPos'].split('/')) >= 2)
# Constraint: At least one Power Forward (PF)
model.PF = pe.Constraint(expr=sum(model.x[i] for i in model.players if 'PF' in feb22_stats.loc[i, 'RosterPos'].split('/')) >= 2)
# Constraint: At least one Center (C)
model.C = pe.Constraint(expr=sum(model.x[i] for i in model.players if 'C' in feb22_stats.loc[i, 'RosterPos'].split('/')) >= 1)
#Utility Constraint - No specific constraint needed since the utility position allows for any of the 5 positions, and it is implicitly fulfilled by the other constraints.
```

**Solve the Model to see if it's Optimal**

```python
opt = pe.SolverFactory('glpk')
result = opt.solve(model)
print(result.solver.status, result.solver.termination_condition)
```

What's our new total DK points given our all our constraints for problem 2?

```python
# Display the selected lineup and total points
selected_lineup = [i for i in model.players if pe.value(model.x[i]) == 1]
total_points = sum(
    scoring_dict[category] * feb22_stats[category][i] if category not in ['Double-Double', 'Triple-Double']
    else scoring_dict[category] * (feb22_stats[category][i] == 'Yes')
    for category in scoring_dict.keys()
    for i in selected_lineup
)

print("Total DK Points: {:.2f}".format(total_points))
```

**Problem 2 Lineup**


Now that we have derived the optimal solution, let's analyze which players stand out as the best performers considering all imposed constraints.






```python
# Extract selected lineup information
selected_lineup_data = feb22_stats.loc[selected_lineup]

# Display the selected lineup in a table
output_table = selected_lineup_data[['Player', 'RosterPos', 'GameInfo', 'PTS', '3P', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'Double-Double', 'Triple-Double', 'SalaryFeb']].copy()
output_table['DK_Points'] = sum(
    scoring_dict[category] * selected_lineup_data[category] if category not in ['Double-Double', 'Triple-Double']
    else scoring_dict[category] * (selected_lineup_data[category] == 'Yes')
    for category in scoring_dict.keys()
)

# Display the table
output_table
```

##Problem 1

To initiate the solution process for problem 1, I began by compiling the Draft Kings (DK) scoring rules outlined in Figure 1 into a dataframe. It's worth noting that the Double-Double and Triple-Double scoring rules were excluded in this dataframe, as we are assuming a lack of knowledge regarding player performance in the Feb 22 games. Further details on the approach taken to tackle this problem will be discussed in the objective function setup section.






```python
# Scoring rules DataFrame
scoring_rules_1 = pd.DataFrame({
    'Category': ['PTS', '3P', 'TRB', 'AST', 'STL', 'BLK', 'TOV'],
    'Points': [1, 0.5, 1.25, 1.5, 2, 2, -0.5]
})
scoring_rules_1
```

```python
# Create a dictionary for easier lookup
scoring_dict_1 = dict(zip(scoring_rules_1['Category'], scoring_rules_1['Points']))
```

**Model Setup and Decison Variables**


```python
# Define the Pyomo Concrete Model for Problem 1
model_1 = pe.ConcreteModel()
```

```python
# Define decision variables for Problem 1
model_1.players = pe.Set(initialize=player_stats.index)
model_1.x = pe.Var(model_1.players, domain=pe.Binary)
for DV in model_1.component_objects(pe.Var):
    DV.pprint()
```

In addressing problem 1, the objective is to construct the optimal starting lineup by evaluating player performance across all DraftKings (DK) categories up to the specified date (Feb 22). This entails considering every DK category except for double-doubles and triple-doubles, as these are bonuses applied per game, while the available PlayerStats data covers all NBA games up to Feb 22. The selection process is guided by identifying players who have excelled within the given constraints and additional criteria:

Exclusion of players from the two lowest-performing NBA teams at present (Detroit Pistons - DET or Washington Wizards - WAS).
Enforcing a minimum Field Goal Percentage (FG%) threshold of above .45 (45%), a benchmark that selects players with an FG%  that meets the league average,
Imposing a minimum requirement of 1000 minutes played throughout the season. Given the average minutes per player per game in the NBA (approximately 20 minutes) multiplied by the number of games played up to Feb 22 (56), this criterion ensures the selection of players who have received substantial playing time, enhancing their potential to accumulate points in games.





```python
# Define the objective function using the scoring dictionary for Problem 1
model_1.obj = pe.Objective(
    expr=sum(
        scoring_dict_1[category] * player_stats[category][i] * model_1.x[i]
        for category in scoring_dict_1.keys()
        for i in model_1.players
    ),
    sense=pe.maximize
)
model_1.obj.pprint()
```

**Define Constraints**

Let's outline the constraints provided in the problem statement as well as the additional constraints introduced in our discussion:

```python
# Define constraints for Problem 1

# We need exactly 8 players
model_1.total_players = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players) == 8)
# The total salary for all the payers has to be less than or equal to 50,000
model_1.total_salary = pe.Constraint(expr=sum(player_stats['SalaryPS'][i] * model_1.x[i] for i in model_1.players) <= 50000)
# We need players from at least 2 games
model_1.games_constraint = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if player_stats['GameInfoPS'][i] != player_stats['GameInfoPS'][selected_lineup[0]]) >= 1)
# We need players in these positions
model_1.PG = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if 'PG' in player_stats.loc[i, 'RosterPS'].split('/')) >= 2)
model_1.SG = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if 'SG' in player_stats.loc[i, 'RosterPS'].split('/')) >= 2)
model_1.SF = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if 'SF' in player_stats.loc[i, 'RosterPS'].split('/')) >= 2)
model_1.PF = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if 'PF' in player_stats.loc[i, 'RosterPS'].split('/')) >= 2)
model_1.C = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if 'C' in player_stats.loc[i, 'RosterPS'].split('/')) >= 1)

# New Additional Constraints for Problem 1
# FG% more than .45
model_1.FG_percentage = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if player_stats.loc[i, 'FG%'] >= 0.45) >= 8)
# No players from DET or WAS
model_1.bad_teams_constraint = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if player_stats['Tm'][i] in ['DET', 'WAS']) == 0)
# Players must have played for more than 1,000 minutes so far this season
model_1.minutes_constraint = pe.Constraint(expr=sum(model_1.x[i] for i in model_1.players if player_stats.loc[i, 'MP'] >= 1000) >= 8)

```

**Solve the Model to see if it's Optimal**

```python
opt = pe.SolverFactory('glpk')
result = opt.solve(model_1)
print(result.solver.status, result.solver.termination_condition)
```

What is the cumulative DK Points tally for the selected players throughout the season, considering the newly imposed constraints?

```python
# Display the selected lineup and total points for Problem 1
selected_lineup_1 = [i for i in model_1.players if pe.value(model_1.x[i]) == 1]
total_points_1 = sum(
    scoring_dict_1[category] * player_stats[category][i] if category not in ['Double-Double', 'Triple-Double']
    else scoring_dict_1[category] * (player_stats[category][i] == 'Yes')
    for category in scoring_dict_1.keys()
    for i in selected_lineup_1
)

print("Total DK Points for the season before Feb 22 games: {:.2f}".format(total_points_1))
```
